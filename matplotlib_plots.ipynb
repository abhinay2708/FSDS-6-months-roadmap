```python
# Import Matplotlib and dependencies
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.figure(figsize=(6,4))

x1=np.linspace(0,10,100)

#create a plot
fig=plt.figure()
plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
plt.show()
```

```python
#Line plot
plt.subplot(2,1,1) # row,column,panel number
plt.plot(x1,np.sin(x1))

plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1))
plt.show()
```

```python
x=np.linspace(0,2,100)

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadretic')
plt.plot(x,x**3,label='cubic')
plt.title('simple plot')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
plt.show()
```

```python
# Scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x=x,y=y)
plt.show()
```

```python
# Histogram
data=np.random.randn(1000)
plt.hist(data)
plt.show()
```

```python
# Barchart
data=[5,50,25,20]
plt.bar(range(len(data)),data)
plt.show()
```

```python
# Error bar chart
x=np.arange(0,4,0.2)
y=np.exp(-x)
e1=0.1*np.abs(np.random.randn(len(y)))
plt.errorbar(x,y,yerr=e1,fmt='.-')
plt.show()
```

```python
# Stacked Bar Chart
a=[15,30,45,22]
b=[15,25,50,20]
c=range(4)
plt.bar(c,a,color='b')
plt.bar(c,b,color='r',bottom=a)
plt.show()
```

```python
# Pie Chart
x=[35,25,20,20]
plt.pie(x)
plt.show()
```

```python
# Box Plot
data=np.random.randn(100)
plt.boxplot(data)
plt.show()
```

```python
# Area Chart
x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y)
plt.show()
```

```python
# Counter Plot
matrix=np.random.rand(10,20)
plt.contour(matrix)
plt.show()
```
