# Evaluation Merics for Regression Models

In regression problems, the target variable is continuous, meaning the model predicts numerical values such as price, temperature, or sales. To understand how well a regression model performs, we use evaluation metrics that measure the difference between the actual values and the predicted values. These metrics help us quantify the accuracy of the model and determine how close the predictions are to the real outcomes. Common regression evaluation metrics include Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), R² Score, and Adjusted R². Each metric provides a different perspective on model performance and helps in selecting the most suitable model for a given problem.


### MAE (Mean Absolute Error):

**MAE (Mean Absolute Error) is a regression evaluation metric used to measure the average difference between the actual values and the predicted values.**

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

Where:
* y_i = actual value
* y^i = predicted value
* n= number of observations

**MAE tells us how far the predictions are from the real values on average.**

- Lower MAE → better model
- MAE = 0 → perfect prediction

**Example:**

| Actual | Predicted | Error | Absolute Error |
| ------ | --------- | ----- | -------------- |
| 10     | 12        | 2     | 2              |
| 20     | 18        | -2    | 2              |
| 30     | 29        | -1    | 1              |

MAE=(2+2+1)/3=1.67

So, the model prediction is on average 1.67 units away from the true value.

#### Why Absolute Value is Used

**We use absolute value so that negative and positive errors do not cancel each other.**

**Example:**

- Error = -3 and +3
- Without absolute → average = 0 (wrong)
- With absolute → average = 3 (correct)

```python
from sklearn.metrics import mean_absolute_error

y_true = [10, 20, 30]
y_pred = [12, 18, 29]

mae = mean_absolute_error(y_true, y_pred)
print(mae)
```

### MSE (Mean Squared Error)

**MSE (Mean Squared Error) is a regression evaluation metric that measures the average of the squared differences between actual values and predicted values.**

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

**MSE tells us how much the predictions differ from the actual values, but the errors are squared.**

- Lower MSE → better model
- 0 → perfect prediction

**Example**

| Actual | Predicted | Error | Squared Error |
| ------ | --------- | ----- | ------------- |
| 10     | 12        | 2     | 4             |
| 20     | 18        | -2    | 4             |
| 30     | 29        | -1    | 1             |

MSE=(4+4+1)/3
MSE=3

So the average squared error is 3.

#### Why Squaring is Used

**Squaring does two things:**

- Removes negative signs
- Penalizes large errors more

**Example:**

| Error | Squared |
| ----- | ------- |
| 1     | 1       |
| 5     | 25      |

Large mistakes get much bigger penalty.

```python
from sklearn.metrics import mean_squared_error

y_true = [10, 20, 30]
y_pred = [12, 18, 29]

mse = mean_squared_error(y_true, y_pred)
print(mse)
```

#### Difference Between MAE and MSE

**MAE calculates the average absolute difference between actual and predicted values, while MSE calculates the average of squared differences. MSE penalizes large errors more heavily than MAE, making it more sensitive to outliers.**

| Feature                 | MAE (Mean Absolute Error) | MSE (Mean Squared Error) |
| ----------------------- | ------------------------- | ------------------------ |
| Formula                 | $\frac{1}{n}\sum(y - \hat{y})$       |   $\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $           |
| Error type              | Absolute error            | Squared error            |
| Penalizes large errors  | No                        | Yes                      |
| Sensitivity to outliers | Low                       | High                     |
| Interpretation          | Average error             | Average squared error    |

#### RMSE (Root Mean Squared Error):

**RMSE (Root Mean Squared Error) is a regression evaluation metric that measures the square root of the average squared differences between actual values and predicted values.**

$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

**Example**

| Actual | Predicted | Error | Squared Error |
| ------ | --------- | ----- | ------------- |
| 10     | 12        | 2     | 4             |
| 20     | 18        | -2    | 4             |
| 30     | 29        | -1    | 1             |

**Step 1: Calculate MSE**

- MSE=(4+4+1)/3=3

**Step 2: Take square root**

- RMSE=3=1.73

So the prediction error is about 1.73 units.

#### Why RMSE is Used

**Problem with MSE:**

- MSE unit becomes squared (e.g., dollars², meters²).

**RMSE solves this because:**

- RMSE has the same unit as the target variable.

**Example:**

- House price prediction → RMSE in rupees
- Temperature prediction → RMSE in °C

```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_true = [10, 20, 30]
y_pred = [12, 18, 29]

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(rmse)
```

### R² Score (Coefficient of Determination)

**R² Score is a regression evaluation metric that measures how well the model explains the variance in the target variable.**

R² tells us how well the regression model fits the data.

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

- $SS_{res}$ = Sum of Squared Residuals (errors from the model)  
- $SS_{tot}$ = Total Sum of Squares (variance in the actual data)

#### Range of R² Score

| R² Value | Meaning                     |
| -------- | --------------------------- |
| **1**    | Perfect model               |
| **0**    | Model predicts same as mean |
| **< 0**  | Model worse than baseline   |

**Example:**

If
R² = 0.85

**Then:**

- 85% variance explained by model
- 15% variance not explained

```python
from sklearn.metrics import r2_score

y_true = [10, 20, 30]
y_pred = [12, 18, 29]

r2 = r2_score(y_true, y_pred)
print(r2)
```

### Adjusted R²

**Adjusted R² is a modified version of R² score that penalizes the model for adding unnecessary features.**

**Adjusted R² tells us how well the model explains the data while considering the number of features used.**

### Why Adjusted R² is Needed

**Problem with R²:**

**When we add more features, the R² value always increases or stays the same, even if the new features are useless.**

- So R² can give a false impression that the model is improving.
- Adjusted R² fixes this problem.

**Formula**
$$
Adjusted\ R^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - p - 1} \right)
$$

Where:

- $R^2$ = R² score  
- $n$ = number of observations  
- $p$ = number of features  

### How It Works

| Situation             | R²        | Adjusted R² |
| --------------------- | --------- | ----------- |
| Useful feature added  | Increases | Increases   |
| Useless feature added | Increases | Decreases   |

So Adjusted R² helps detect unnecessary variables.

**Example**

Suppose:

**Model 1**

- Features = 2
- R² = 0.82
- Adjusted R² = 0.81

**Model 2**

- Features = 10
- R² = 0.84
- Adjusted R² = 0.79

Even though R² increased, Adjusted R² decreased, meaning the new features are not useful.
