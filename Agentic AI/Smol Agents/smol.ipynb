```python
model= LiteLLMModel(
    model_id= "gemini/gemini-2.5-flash-lite-preview-06-17",
    api_key= os.getenv("GOOGLE_API_KEY")
)
```

```python
ml_code_agent = CodeAgent(
    tools = [DuckDuckGoSearchTool()],
    additional_authorized_imports= ['pandas', 'numpy', 'sklearn', 'json'],
    model=model
)
```

```python
def generate_ml_code(prompt):
    try:
        response = ml_code_agent.run(prompt)
        return response
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
iface = gr.Interface(
    fn = generate_ml_code,
    inputs='text',
    outputs='text',
    title=' ML code Generator',
    description='Enter a Machine Learning Coding Task'
)
```

```python
iface.launch()
```
